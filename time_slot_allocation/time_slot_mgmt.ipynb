```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./nda_list.xlsx').query('Confirmed == "Yes"')
```

```python

df_test = df.copy()
```

```python
hours_range = np.arange(0,24, .5)
```

```python
list_ = []
for hour_ in hours_range:

    attendees = (df.Timezone + hour_ > 8) & (df.Timezone + hour_ < 17)
    total_first_attendees = sum(attendees)

    for hour_2 in hours_range[hours_range!=hour_]:
        attendees_2  = (df[~attendees].Timezone + hour_2 > 8) & (df[~attendees].Timezone + hour_2 < 17)
        total_second_attendees = sum(attendees_2)

        total_attendess = total_first_attendees + total_second_attendees
    

        attendees_t  = (df.Timezone + hour_2 > 8) & (df.Timezone + hour_2 < 17)
        total_t_attendees = sum(attendees_t)

        total_attendess_t = total_first_attendees + total_t_attendees


        list_.append((str(hour_), total_first_attendees, str(hour_2), total_second_attendees, total_t_attendees, total_attendess, total_attendess_t))
```

```python
time_slots= pd.DataFrame(list_, columns=['hour_1', 'attendees_1', 'hour_2', 'attendees_2', 'attendees_2_not_excluding', 'total', 'total_total'])
```

```python
time_slots.sort_values([ 'total', 'total_total','attendees_1', 'attendees_2'], ascending=[False, False, False, False]).head(50)
```

```python
time_slots['sum'] = time_slots.hour_1.apply(lambda x: 2**float(x)) + time_slots.hour_2.apply(lambda x: 2**float(x))
```

```python
time_slots['overlap'] = time_slots.total_total -  time_slots.total
```

```python
time_slots.drop_duplicates('sum', keep='last').sort_values([ 'total', 'total_total','attendees_1', 'attendees_2'], ascending=[False, False, False, False]).head(50)
```
